```julia
using Astrodynamics
using DerIMDEX
using ODE
using jspice
using PyPlot
```

```julia
jsinit()
```

```julia
mu=planets["earth"]["mu"]
RE=planets["earth"]["re"]
hp=185
inc=28.5*pi/180
e=0.975000919
node=86.83224*pi/180
argp=-129.98616*pi/180
jd0=2458119.33552
dt=5.23359
dt2=5.3
rp=hp+RE
a=rp/(1-e)
```

```julia
el=[a;e;inc;node;argp;0]
x0=cartesian(el,mu)
```

```julia
gregorian(jd0,"jd")
```

```julia
et0=(jd0-julian(2000,1,1,12,0,0,"jd"))*86400;
et1=et0+dt2*86400;
```

```julia
traj=ode45(der,x0,[et0;et1]);
n=length(traj[1]);
x=[traj[2][j][1] for j=1:n];
y=[traj[2][j][2] for j=1:n];
z=[traj[2][j][3] for j=1:n];
s_eci=[traj[2][j][1:6] for j=1:n];
```

```julia
s_eci[3]
```

```julia
t_mjd2k=traj[1]/86400+0.5;
```

```julia
eci2rotpulse(s_eci[1],t_mjd2k[1])
```

```julia
s_rot=[eci2rotpulse(s_eci[j],t_mjd2k[j]) for j=1:n]
```

```julia
frlin=plot3D(x/1e4,y/1e4,z/1e4);
axis("equal");
xlabel("\$x\\ [10^4 {\\rm km}]\$");
ylabel("\$y\\ [10^4 {\\rm km}]\$");
zlabel("\$z\\ [10^4 {\\rm km}]\$");
grid(true);
```

```julia
s_rot
```

```julia
xr=[s_rot[j][1] for j=1:n];
yr=[s_rot[j][2] for j=1:n];
zr=[s_rot[j][3] for j=1:n];
vxr=[s_rot[j][4] for j=1:n];
vyr=[s_rot[j][5] for j=1:n];
vzr=[s_rot[j][6] for j=1:n];
```

```julia
frlin=plot(xr,yr);
axis("equal");
xlabel("\$x\\ [DU]\$");
ylabel("\$y\\ [DU]\$");
grid(true);
[arrow(xr[j],yr[j],vxr[j]/20,vyr[j]/20) for j=1:n];
```
